# HW4 — W4111 ClassicModels API Demo

This notebook demonstrates all implemented endpoints against the **live deployed API**.

**Live API:** https://template-mauve-eight.vercel.app  
**Interactive Docs:** https://template-mauve-eight.vercel.app/docs

All cells are runnable top-to-bottom with no local server required.

```python
import httpx

BASE = "https://template-mauve-eight.vercel.app"

def show(resp: httpx.Response):
    print(f"{resp.request.method} {resp.url}  →  {resp.status_code}")
    try:
        print(resp.json())
    except Exception:
        print(resp.text)
    print()
```

## Health check

```python
show(httpx.get(f"{BASE}/health"))
```

## Customers

```python
# GET /customers — list all customers
show(httpx.get(f"{BASE}/customers"))
```

```python
# GET /customers?country=USA — filter by country
show(httpx.get(f"{BASE}/customers", params={"country": "USA"}))
```

```python
# GET /customers/103 — retrieve a single customer
show(httpx.get(f"{BASE}/customers/103"))
```

```python
# POST /customers — create a new customer
new_customer = {
    "customerNumber": 99999,
    "customerName": "Test Customer",
    "contactLastName": "Doe",
    "contactFirstName": "Jane",
    "phone": "555-1234",
    "addressLine1": "123 Test St",
    "city": "New York",
    "country": "USA",
    "creditLimit": 5000.00
}
show(httpx.post(f"{BASE}/customers", json=new_customer))
```

```python
# GET /customers/99999 — verify creation
show(httpx.get(f"{BASE}/customers/99999"))
```

```python
# PUT /customers/99999 — update the customer
updated_customer = {
    "customerNumber": 99999,
    "customerName": "Test Customer Updated",
    "contactLastName": "Doe",
    "contactFirstName": "Jane",
    "phone": "555-9999",
    "addressLine1": "456 New Ave",
    "city": "Boston",
    "country": "USA",
    "creditLimit": 10000.00
}
show(httpx.put(f"{BASE}/customers/99999", json=updated_customer))
```

```python
# DELETE /customers/99999 — delete the test customer
show(httpx.delete(f"{BASE}/customers/99999"))
```

```python
# GET /customers/99999 — should return 404 now
show(httpx.get(f"{BASE}/customers/99999"))
```

## Orders

```python
# GET /orders — list all orders
show(httpx.get(f"{BASE}/orders"))
```

```python
# GET /orders?status=Shipped — filter by status
show(httpx.get(f"{BASE}/orders", params={"status": "Shipped"}))
```

```python
# GET /orders/10100 — retrieve a specific order
show(httpx.get(f"{BASE}/orders/10100"))
```

```python
# POST /orders — create a new order (customer 103 must exist)
new_order = {
    "orderNumber": 99998,
    "orderDate": "2024-01-15",
    "requiredDate": "2024-01-30",
    "status": "In Process",
    "customerNumber": 103
}
show(httpx.post(f"{BASE}/orders", json=new_order))
```

```python
# PUT /orders/99998 — update the order
updated_order = {
    "orderNumber": 99998,
    "orderDate": "2024-01-15",
    "requiredDate": "2024-02-05",
    "status": "Shipped",
    "customerNumber": 103
}
show(httpx.put(f"{BASE}/orders/99998", json=updated_order))
```

```python
# DELETE /orders/99998
show(httpx.delete(f"{BASE}/orders/99998"))
```

## Order Details

```python
# GET /orderdetails — list all order details (may be large)
resp = httpx.get(f"{BASE}/orderdetails", params={"orderNumber": 10100})
show(resp)
```

```python
# GET /orders/10100/orderdetails — all details for order 10100
show(httpx.get(f"{BASE}/orders/10100/orderdetails"))
```

```python
# GET /orders/10100/orderdetails/S18_1749 — single order detail
show(httpx.get(f"{BASE}/orders/10100/orderdetails/S18_1749"))
```

```python
# PUT /orders/10100/orderdetails/S18_1749 — update quantity
updated_detail = {
    "orderNumber": 10100,
    "productCode": "S18_1749",
    "quantityOrdered": 50,
    "priceEach": 136.00,
    "orderLineNumber": 3
}
show(httpx.put(f"{BASE}/orders/10100/orderdetails/S18_1749", json=updated_detail))
```

```python
# POST /orderdetails — create a new order detail (requires valid order + product)
# First re-create order 99998 so we have a valid FK
httpx.post(f"{BASE}/orders", json={
    "orderNumber": 99998,
    "orderDate": "2024-01-15",
    "requiredDate": "2024-01-30",
    "status": "In Process",
    "customerNumber": 103
})

new_detail = {
    "orderNumber": 99998,
    "productCode": "S18_1749",
    "quantityOrdered": 10,
    "priceEach": 136.00,
    "orderLineNumber": 1
}
show(httpx.post(f"{BASE}/orderdetails", json=new_detail))
```

```python
# DELETE the new detail, then the order
show(httpx.delete(f"{BASE}/orders/99998/orderdetails/S18_1749"))
show(httpx.delete(f"{BASE}/orders/99998"))
```

## Error cases

```python
# 404 — customer that doesn't exist
show(httpx.get(f"{BASE}/customers/0"))

# 404 — order that doesn't exist
show(httpx.get(f"{BASE}/orders/0"))

# 404 — order detail that doesn't exist
show(httpx.get(f"{BASE}/orders/0/orderdetails/INVALID"))
```
